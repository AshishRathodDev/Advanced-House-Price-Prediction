### Missing Values Handling - Categorical Features

```python
# Identify categorical features with missing values

features_nan=[feature for feature in df.columns if df[feature].isnull().sum()>1 and df[feature].dtypes=='O']

# Print missing value percentages

for feature in features_nan:
    print("{}: {}% missing values".format(feature,np.round(df[feature].isnull().mean(),4)))
```

```python
# Function to replace missing categorical values

def replace_cat_feature(df,features_nan):
    data=df.copy()
    data[features_nan]=data[features_nan].fillna("Missing")
    return data

# Apply the missing value replacement function

df=replace_cat_feature(df,features_nan)
df[features_nan].isnull().sum()
```

```python
# Identify numerical features with missing values

numerical_with_nan=[feature for feature in df.columns if df[feature].isnull().sum()>1 and df[feature].dtypes!='O']

# Print numerical missing value percentages

for feature in numerical_with_nan:
    print("{}: {}% missing value".format(feature,np.around(df[feature].isnull().mean(),4)))
```

```python
# Handling Missing Numerical Values
# Strategy: 
# 1. Replace missing values with median
# 2. Create a new binary feature to indicate missingness

for feature in numerical_with_nan:
    # Calculate median value for imputation
    median_value=df[feature].median()
    
    # Create a new feature to capture whether the original value was missing
    # 1 indicates missing, 0 indicates present
    
    df[feature+'nan']=np.where(df[feature].isnull(),1,0)
    # Fill missing values with median
    df[feature].fillna(median_value,inplace=True)

df[numerical_with_nan].isnull().sum()
```

### Numerical Variable Transformation

```python
# Apply log transformation to handle skewed distributions
# Log transformation helps in:
# 1. Reducing skewness
# 2. Handling exponential growth patterns
# 3. Normalizing highly variable features

```

```python
num_features=['LotFrontage', 'LotArea','1stFlrSF','GrLivArea','SalePrice']
for feature in num_features:
    df[feature]=np.log(df[feature])
```

### Handling Rare Categorical Features

```python
# Identify categorical features
```

```python
# Replace categories that appear in less than 1% of observations
# This helps reduce noise and complexity in categorical variables

for feature in categorical_features:
    temp=df.groupby(feature)['SalePrice'].count()/len(df)
    temp_df1=temp[temp>0.01].index
    df[feature]=np.where(df[feature].isin(temp_df1),df[feature],'Rare_var')
```

```python
# Ordinal Encoding for Categorical Features
# Replace categories with their mean sale price order
# This transforms categorical variables into numerical representation

for feature in categorical_features:
    labels_ordered=df.groupby([feature])['SalePrice'].mean().sort_values().index
    labels_ordered={k:i for i,k in enumerate(labels_ordered,0)}
    df[feature]=df[feature].map(labels_ordered)
```
